```python
import pandas as pd
```

```python
#Прочитайте CSV файл (використовуйте метод read_csv)

# Перелік можливих кодувань
possible_encodings = ['utf-8', 'latin-1', 'cp1251']

# Цикл, який перебирає всі можливі кодування
for encoding in possible_encodings:
    try:
        # Спроба зчитати файл з поточним кодуванням
        data = pd.read_csv(r'C:\Users\admin\Desktop\Global YouTube Statistics.csv', encoding=encoding)
        print("Файл успішно зчитано з кодуванням:", encoding)
        # Якщо вдалося зчитати, вийти з циклу
        break
    except UnicodeDecodeError:
        print("Не вдалося зчитати з кодуванням:", encoding)
```

```python
#Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків  у кожному зі стовпців (використовуйте методи isna та sum). 
#Якщо є пропуски, замінить їх на NaN, потім тип даних на числовий. 

print('Кількість пропусків у кожному зі стовпців: ', data.isna().sum())

data_NaN = data.fillna('NaN')

data_NaN.replace('NaN', 0)

print(data_NaN)
```

```python
#Замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float. 

data_mean = data.replace(0, data.mean())
data_mean.apply(pd.to_numeric, errors='coerce')
```

```python
#Ще раз перевірте кількість пропусків.

print('Кількість пропусків: ', data.isna().sum())
```

```python
#Перевірте, скільки унікальних країн представлено у колонці "Country" 

print("Кількість унікальних країн:", data['Country'].nunique())
```

```python
# Побудуйте діаграму розподілу переглядів (використовуйте kind='hist'):

import matplotlib.pyplot as plt

data['video views'].plot(kind = 'hist')
plt.title('Розподіл переглядів')

plt.show()
```

```python
#Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean) 

data_max = data['video views'].max()
data_min = data['video views'].min()
data_mean = data['video views'].mean()

print("Максимальна кількість переглядів:", data_max)
print("Мінімальна кількість переглядів:", data_min)
print("Середня кількість переглядів:", data_mean)
```

```python
#Знайдіть країну, де найбільше відео було завантажено на YouTube

country = data.sort_values('uploads', ascending = False)
max_country = country.head(1)
print(max_country['Country'])
```

```python
#Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads. 

video = data.sort_values('uploads', ascending = False)
max_count = video.head(1)
print(max_count['Title'])
min_count = video.tail(1)
print(min_count['Title'])
```
